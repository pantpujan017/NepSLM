## Load Gemma4 Model

```python
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load Gemma-4-E2B-it
model_id = "google/gemma-4-E2B-it"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16,      # better than float16 for Gemma-4
    device_map="auto",
    low_cpu_mem_usage=True,
    attn_implementation="sdpa"       # recommended for speed
)

print("Model loaded successfully!")
```

```python
import math
from tqdm import tqdm

# ================== Correct Perplexity for Gemma-4-it ==================
def calculate_perplexity_gemma4(texts, model, tokenizer, max_samples=1012):
    model.eval()
    total_loss = 0.0
    total_tokens = 0

    for text in tqdm(texts[:max_samples], desc="Computing PPL"):
        if len(text.strip()) < 10:
            continue

        # Use proper chat template (this is the key fix)
        messages = [
            {"role": "user", "content": f"Continue this Nepali sentence naturally and fluently:\n{text}"}
        ]

        prompt = tokenizer.apply_chat_template(
            messages,
            tokenize=False,
            add_generation_prompt=True
        )

        inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

        if inputs.input_ids.size(1) < 20:  # avoid too short inputs
            continue

        with torch.no_grad():
            outputs = model(**inputs, labels=inputs.input_ids)

        loss = outputs.loss.item()
        num_tokens = inputs.input_ids.size(1)

        total_loss += loss * num_tokens
        total_tokens += num_tokens

    if total_tokens == 0:
        return float("inf")

    avg_loss = total_loss / total_tokens
    perplexity = math.exp(avg_loss)
    return perplexity


# Run the corrected function
ppl = calculate_perplexity_gemma4(eval_texts, model, tokenizer, max_samples=1012)
print(f"\n✅ Perplexity on FLORES-200 Nepali (with chat template): {ppl:.2f}")
```

## Devanagri counts

```python
import re
devanagari_pattern = re.compile(r'[\u0900-\u097F]')
```

```python
import numpy as np
import re

def calculate_script_density(example):
    text = example.get('text', example.get('Article', ''))
    if not text:
        return {'deva_ratio': 0.0}

    # Count Devanagari characters
    deva_chars = len(re.findall(devanagari_pattern, text))
    total_chars = len(text.replace(' ', '')) # Ignore whitespace in noise calculation

    ratio = deva_chars / total_chars if total_chars > 0 else 0
    return {'deva_ratio': ratio}

for name, data in [('Wiki', wiki), ('IRIS', iris)]:
    # Sample documents for a quick assessment
    sample_size = min(5000, len(data))
    sample_data = data.shuffle(seed=42).select(range(sample_size))

    results = sample_data.map(calculate_script_density)
    avg_density = np.mean(results['deva_ratio'])

    print(f"--- {name} Dataset (Sample: {sample_size}) ---")
    print(f"Average Nepali Script Density: {avg_density:.2%}")
    print(f"Noise/Other Content: {1 - avg_density:.2%}\n")
```

## Calculate Average Length


```python
import numpy as np

def calculate_length_stats(example):
    text = example.get('text', example.get('Article', ''))
    if not text:
        return {'char_count': 0, 'word_count': 0}
    return {
        'char_count': len(text),
        'word_count': len(text.split())
    }

for name, data in [('Wiki', wiki), ('IRIS', iris)]:
    # Use a sample for speed or process full if small
    sample_size = min(10000, len(data))
    sample_data = data.shuffle(seed=42).select(range(sample_size))

    stats = sample_data.map(calculate_length_stats)

    avg_chars = np.mean(stats['char_count'])
    median_chars = np.median(stats['char_count'])
    avg_words = np.mean(stats['word_count'])
    median_words = np.median(stats['word_count'])

    print(f"--- {name} Dataset Length Analysis ---")
    print(f"Average Characters: {avg_chars:.2f} (Median: {median_chars})")
    print(f"Average Words: {avg_words:.2f} (Median: {median_words})\n")
```

## Combine Data + Dedup

```python
from datasets import concatenate_datasets, Dataset

# Normalize to same column name — just 'text'
iriis = iris.rename_column("Article", "text")
iriis = iriis.remove_columns([c for c in iriis.column_names if c != "text"])
# Already 'text'
wiki = wiki.remove_columns([c for c in wiki.column_names if c != "text"])

# Combine
combined = concatenate_datasets([iriis, wiki])
print(f"Total docs: {len(combined)}")

# Check the data
print(combined[0])
print(combined[-1])

# Save
# combined.to_parquet("./nepali_pretrain_raw.parquet")
```

```python
# ── 2. EXACT DEDUP (in memory, nearly free) ───────────────
seen = set()
def is_unique(example):
    h = hash(example["text"].strip().lower())
    if h in seen:
        return False
    seen.add(h)
    return True

combined = combined.filter(is_unique, num_proc=1)  # must be 1 for shared set
print(f"After exact dedup: {len(combined)}")
```

```python
# check CPU Cores
import os
print(os.cpu_count())
cpu_count = os.cpu_count()
```

```python
import unicodedata
import re
# ── 3. CLEAN (in memory) ──────────────────────────────────
def clean(example):
    text = unicodedata.normalize("NFC", example["text"]).strip()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    if len(text) < 50 or len(text) > 100_000:
        return {"text": None}

    deva  = len(re.findall(r'[\u0900-\u097F]', text))
    total = len(text.replace(' ', ''))
    if total > 0 and deva / total < 0.5:
        return {"text": None}

    return {"text": text}

combined = combined.map(clean, num_proc=cpu_count)
combined = combined.filter(lambda x: x["text"] is not None, num_proc=cpu_count)
print(f"After cleaning: {len(combined)}")
```

```python
# ── 4. NEAR DEDUP ─────────────────────────────────────────
from datasketch import MinHash, MinHashLSH

lsh  = MinHashLSH(threshold=0.85, num_perm=128)
keep = []

for i, example in enumerate(combined):
    m = MinHash(num_perm=128)
    for w in example["text"].split():
        m.update(w.encode("utf-8"))
    if not lsh.query(m):
        lsh.insert(f"d{i}", m)
        keep.append(True)
    else:
        keep.append(False)

combined = combined.filter(lambda _, idx: keep[idx], with_indices=True)
print(f"After near dedup: {len(combined)}")
```
